```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Load the data
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data_preprocessed
```

```python
#Targets
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head(10)
```

```python
targets.sum()/targets.shape[0]
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'], axis=1)
```

```python
data_with_targets is data_preprocessed
```

```python
data_with_targets.head()
```

```python
#Selecting the inputs for the logistic regression
data_with_targets.shape
```

```python
unscaled_inputs = data_with_targets.iloc[:, :-1]
unscaled_inputs
```

```python
#Standardizing the data
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs.shape
```

```python
#Splitting the date for training and testing
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs, targets)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, shuffle=True, random_state=20)
```

```python
y_test.shape
```

```python
#Fitting the Model and Assessing it's accuracy
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
##training the model
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
##Manually check the accuracy
model_outputs = reg.predict(x_train)
model_outputs
```

```python
np.sum(model_outputs == y_train)/model_outputs.shape[0]  #the accuracy from the model is correct!
```

```python
#Creating a Summary Table with the coefficient and Intercept
reg.intercept_
```

```python
reg.coef_
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'], data=feature_name)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
##Adding interception at the first place
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
#Interpreting the coefficients for our Problem
summary_table['odds_ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

```python
summary_table.sort_values('odds_ratio', ascending=False)
```
