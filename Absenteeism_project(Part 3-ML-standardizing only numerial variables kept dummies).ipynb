```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy, with_mean, with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

1. important values 3,1,2,4,5,11,9 & 10 Education,12 Pets

3. 5 reasons variable, 0-no reason(baseline model-when no reason is given), 1-various diseases, 2-pregnacy and giving birth, 3-poisioning, 4-light diseases
*poisoning: The weight means the odds of someone being excessively absent after being poisoned are 20 times higher than when no reason was reported.
*various diseases: the normal absenteeism case, you got sick, you skipped work. A person who has reported this is 14 times more likely to be excessively absent than a person who didn't specify a reason
*Pregnacy and giving birth: pregnacy check and come back to work, from time to time, there's some emergencies, but the odds ratio is only 2.3 times more likely to be excessively absent than the base model.

4. Transportation Expense: is one of our standardized variables, we don't have direct interpretability of it. the odds ratio implies that for 1 standardized unit or for 1 standard deviation increase in transportation expense, it is close to twice as likely to be excessively absent
5. Pet: for each additional standardized unit of pet, the odds are 1-0.751567 = 25% lower than the base model(no pet) explanation: maybe if you have many pets, you are probably not taking care of them on your own, you will have someone to take pets to the doctor if something is wrong.
